```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pprint
%matplotlib inline
```

```python
x = np.arange(0,25)
y = (2 * x) ** 2 + 5
z = x * .5
print(x, y, z)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(y,z)
ax.set_xlabel('y')
ax.set_ylabel('z')
ax.set_title('y vs z')
plt.show
```
